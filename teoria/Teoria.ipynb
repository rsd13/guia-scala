# Introduccin

Scala es un lenguaje de programacin de uso general que viene con soporte de programacin funcional y un fuerte sistema de tipo esttico

El cdigo fuente de Scala est destinado a ser compilado en Java bytecode, para que el cdigo ejecutable resultante se pueda ejecutar en una Java virtual machine (JVM).




- **Scala est orientada a objetos**
    - Scala es un muy buen ejemplo de un lenguaje orientado a objetos .
    - Se usan clases y traits (interfaces)
   
- **Scala es functional**
    - La programacin funcional trata funciones como otros componentes del lenguaje
    - Scala usa una forma sencilla de definir funciones annimas (funciones sin nombres).
    - Tambin es compatible con higher order functions (funciones de primera clase) que permite crear funciones anidadas.
- **Scala es statically typed**
    - A diferencia de los otros lenguajes escritos estticamente como Pascal, Scala no es necesario indicar el tipo, lo infiere en la mayora de los casos.
    - Por tanto todas las comprobaciones de tipos las realiza el compilador
    
- **Scala es statically typed**
    - Se pueden usar clases de Java SDK
    - Se pueden ejecutar programas realizados en Java
- **Scala puede hacer procesos concurrentes y sincronizados

## Diferencias con Java
- **Todos los tipos son objetos**
    - Todo es un como un objeto en Scala, aunque a veces su implementacin es interna
    - Por ejemplo, en Scala, las cadenas se convierten implcitamente en colecciones de caracteres, pero no en Java
- **Inferencia de tipos**
    - Deduccin de tipos en tiempo de compilacin. (no hay que poner el tipo de objeto)
    - No en RUNTIME
- **Scala REPL**
    - Herramienta en modo comando para trabajar y probar comandos Scala
- **Funciones anidadas**

- **Imports**
    - En Java, solo puede importar paquetes en la parte superior del cdigo
    - En Scala se pueden hacer imports casi en cualquier parte por ejemplo, dentro de una clase o un mtodo).
    - El _ (subrayado) en Scala se utiliza para las importaciones de forma similar al * (asterisco) que se utilizara en java
- **Mayor legibilidad y sencillez en el cdigo**
- **Tiene una mayor curva de aprendizaje que Java**
- **Se pueden sobrecargar operadores**
